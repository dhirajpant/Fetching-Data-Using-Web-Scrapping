```python
#importing necessary libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#extracting and storing the raw HTML content of the webpage in the variable abc located at the given URL.
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
abc = requests.get('https://www.gizmochina.com/phones/page/1/',headers=headers).text
```

```python
abc
```

```python
sp = BeautifulSoup(abc, 'lxml')
```

```python
#prints out the HTML content with well-formatted indentation
print(sp.prettify)
```

```python
#extract all the <h2> elements from the parsed HTML represented by the sp object.
sp.find_all('h2')
```

```python
#extract and returns a list of all the <div> elements with the class 'aps-product-box' from the parsed HTML represented by the sp object
product=sp.find_all('div',class_='aps-product-box')
```

```python
name=[]  #Initialize an empty list to store product names
Rating=[] # Initialize an empty list to store ratings
for i in product:
    name.append(i.find('h2').text)
    Rating.append(i.find('span',class_='aps-comp-rating').text)
```

```python
name
```

```python
Rating
```

```python
d={'name':name,'Rating':Rating}
```

```python
df=pd.DataFrame(d)
```
